```python
def getDemo(demo, field):
    clusteredData = pd.DataFrame(columns=['specifics','company','variable','value'])
    for empresai in demo:
        aux = pd.DataFrame.copy(empresai[field])
        aux['specifics'] = aux.index
        aux['company'] = empresai['nomeDaEmpresa'][1]
        clusteredData= pd.concat((clusteredData,aux.melt(id_vars=['specifics','company'])),axis=0)
    return clusteredData[clusteredData['value'].isnull() == False]
```

```python
def sumToTotal(matrix,gender):
    eachOfGender = ['mulheres' in x.lower() and 'total' not in x.lower()  for x in list(matrix.columns)]
    totalGender = ['mulheres' in x.lower() and 'total' in x.lower() for x in list(matrix.columns)]
    eachSum =  matrix.iloc[:,eachOfGender].sum(axis=1)
    total = matrix.iloc[:,totalGender].sum(axis=1)
    return np.array(eachSum, dtype=np.float), np.array(total, dtype=np.float)

def compareGenderSum(matrix,gender):
    each,total = sumToTotal(matrix,gender)
    return np.logical_or( (each==total), np.logical_and(noValue(each),noValue(total) ) )

def noValue(array):
    return np.logical_or(np.isnan(array),array==0)

def isDiscriminated(matrix):
    sumH, _ = sumToTotal(matrix,'homens')
    sumM, _ = sumToTotal(matrix,'mulheres')
    return sumH>0, sumM>0
```
